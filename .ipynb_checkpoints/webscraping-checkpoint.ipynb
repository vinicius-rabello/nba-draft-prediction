```python
import requests as req
from bs4 import BeautifulSoup
import numpy as np
from tqdm import tqdm
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
import duckdb
```

### Creating NCAA stats dataframe

```python
classes=['Fr','So','Jr','Sr']
years=np.arange(2003,2025,1)
columns=['player','cls','year','gp','mpg','ppg','fgm','fga','fg%','3pm','3pa','3p%','ftm','fta','ft%','orb',
        'drb','rpg','apg','spg','bpg','tov','pf']
```

```python
table=[]
for year in tqdm(years):
    for cls in classes:
        for page in range(1,11):
            try:
                url=f'https://basketball.realgm.com/ncaa/stats/{year}/Averages/Qualified/{cls}/Season/All/points/desc/{page}'
                r=req.get(url)
                soup=BeautifulSoup(r.text)
                for i in range(1,31):
                    row=[str(soup.find_all('tr')[i]).split('</td>')[1].split('>')[-2][:-3],cls,year]
                    for j in range(3,23):
                        row.append(str(soup.find_all('tr')[i]).split('</td>')[j][4:])
                    table.append(row)
            except:
                print('error: '+url)
                pass
```

```python
df=pd.DataFrame(table,columns=columns)
```

```python
df.to_excel('ncaa-stats-complete.xlsx',index=False)
```

```python
df
```

### Creating Advanced Stats Dataset

```python
classes=['Fr','So','Jr','Sr']
years=np.arange(2003,2025,1)
columns=['player','cls','year','TS%','eFG%','totalS%','orb%','drb%','trb%','ast%','tov%','stl%','blk%',
        'usg%','ppr','pps','ORtg','DRtg','eDiff','FIC','PER']
```

```python
table=[]
for year in tqdm(years):
    for cls in classes:
        for page in range(1,11):
            try:
                url=f'https://basketball.realgm.com/ncaa/stats/{year}/Advanced_Stats/Qualified/{cls}/Season/All/per/desc/{page}'
                r=req.get(url)
                soup=BeautifulSoup(r.text)
                for i in range(1,31):
                    row=[str(soup.find_all('tr')[i]).split('</td>')[1].split('>')[-2][:-3],cls,year]
                    for j in range(3,21):
                        row.append(str(soup.find_all('tr')[i]).split('</td>')[j][4:])
                    table.append(row)
            except:
                print('error: '+url)
                pass
```

```python
df=pd.DataFrame(table,columns=columns)
```

```python
df.to_excel('ncaa-advanced-stats-complete.xlsx',index=False)
```

```python
df
```

### Creating NBA drafts dataframe 

```python
years=np.arange(2003,2024,1)
```

```python
table=[]

for year in tqdm(years):
    try:
        url=f'https://basketball.realgm.com/nba/draft/past_drafts/{year}'
        r=req.get(url)
        soup=BeautifulSoup(r.text)
        rows=soup.find_all('table')[4].find('tbody') # fourth table is the first round draft
        for row in rows:
            entry=[]
            if row=='\n':
                continue
            else:
                for i in range(12):
                    entry.append(row.find_all('td')[i].text)
                    
            entry.append('first round')
            entry.append(year)
            table.append(entry)
    except:
        print('error: '+url)
        
for year in tqdm(years):
    try:
        url=f'https://basketball.realgm.com/nba/draft/past_drafts/{year}'
        r=req.get(url)
        soup=BeautifulSoup(r.text)
        rows=soup.find_all('table')[5].find('tbody') # fifth table is the second round draft
        for row in rows:
            entry=[]
            if row=='\n':
                continue
            else:
                for i in range(12):
                    entry.append(row.find_all('td')[i].text)
                    
            entry.append('second round')
            entry.append(year)
            table.append(entry)
    except:
        print('error: '+url)
```

```python
columns=['pick','player','team','draft_trades','pos','height','weight','age','yos','pre_draft_team',\
         'cls','nationality','round','year']

df=pd.DataFrame(table,columns=columns)
df['pick'] = df['pick'].apply(pd.to_numeric)
```

```python
df=duckdb.query("SELECT * FROM df ORDER BY year,round,pick").df()
df
```

```python
df.to_excel('drafts.xlsx',index=False)
```

### Joining tables

```python
stats=pd.read_excel('ncaa-stats-complete.xlsx')
advanced_stats=pd.read_excel('ncaa-advanced-stats-complete.xlsx')
drafts=pd.read_excel('drafts.xlsx')
```

```python
stats.shape
```

```python
conn = duckdb.connect(':memory:')

conn.register("stats", stats)
conn.register("advancedStats", advanced_stats)
conn.register("drafts", drafts)
```

```python
query = """
    SELECT stats.*,advancedStats.* FROM stats
    INNER JOIN advancedStats ON stats.player = advancedStats.player
    AND stats.year = advancedStats.year
"""
```

```python
df = conn.execute(query).fetchdf()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(columns=['player_1','cls_1','year_1'],inplace=True)
```

```python
conn.register("fullStats", df)
```

```python
query = """
    SELECT fullStats.*,drafts.pick AS draft_pos,drafts.round AS drafted FROM fullStats
    LEFT JOIN drafts ON fullStats.player = drafts.player
    AND fullStats.year = drafts.year
"""
```

```python
df = conn.execute(query).fetchdf()
```

```python
df.to_excel('full_table.xlsx',index=False)
```

```python
df=pd.read_excel('full_table.xlsx')
```
